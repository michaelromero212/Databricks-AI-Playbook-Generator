# Kafka to Bronze Ingestion
Reads streaming data from Kafka topics and writes to Bronze Delta tables.

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *

# Read from Kafka
df_kafka = spark.readStream.format("kafka") \
  .option("kafka.bootstrap.servers", "host1:port1,host2:port2") \
  .option("subscribe", "transactions, customers") \
  .load()

# Write to Bronze
df_kafka.writeStream \
  .format("delta") \
  .outputMode("append") \
  .option("checkpointLocation", "/mnt/delta/bronze/_checkpoints/kafka_ingest") \
  .table("bronze_kafka_raw")
```
